```python
#melihat jumlah imigran dari jepang untuk tahun 2013

print(df_can.loc['Japan', 2013])

print(df_can.iloc[87,36])
```

```python
#melihat jumlah imigran dari jepang untuk tahun 1980 sd 1985
print(df_can.loc['Japan', [1980,1981,1982,1983,1984,1985]])

print(df_can.iloc[87,[3,4,5,6,7,8]])
```

```python
#nama kolom yang merupakan bilangan bulan seperti tahun mungkin menimbulkan kebinganan. 
#misalnya ketiga kita mengacu pada tahun 2013, orang mungkin bingung ketika 2013 adalah idneks posisi 2013

#untuk menghindari ambuitas ini, mari kita ubah namakolom nejadi string '1980' hingga '2013'

df_can.columns = list(map(str, df_can.columns))
```

```python
#karena tahun sudah dikonversi menjadi string,
#maka tahun bisa dijadikan sebagai variabel untuk memudahkan memanggil rentang tahun penuh

years=list(map(str, range(1980,2014)))
years
```

## Filtering based on a Criteria

```python
#1. create the condition boolean series

condition = df_can['Continent'] == 'Asia'
print(condition)
```

```python
# 2. pass this condition into the DataFrame

df_can[condition]
```

```python
# multiple criteria in the same line

df_can[(df_can['Continent']=='Asia') & (df_can['Region']=='Southern Asia') & (df_can['DevName']=='Developing regions')]

# dont forget to put any condition in parentheses
```

# Line Plots

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
df_can.plot()
```

```python
haiti.index=haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.text(2000,6000, '2010 Earthquake')

plt.show()
```

```python
China_India = df_can.loc[['China', 'India'], years]
China_India
```

```python
China_India.plot(kind='line')
```

```python
df_CI=China_India.T
```

```python
df_CI.head()
```

```python
df_CI.plot(kind='line')
plt.title('Immigrants from China and India')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
#compare top 5 countries

df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df_top5 = df_can.head(5)

df_top5 = df_top5[years].transpose()

df_top5.index =df_top5.index.map(int)

df_top5.plot(kind='line', figsize=(14,8))

plt.title('Immigrants from Top Five Countries')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.show()
```
