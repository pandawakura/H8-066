## Unexpected Missing Values

```python
df['OWN_OCCUPIED']

# index 6 merupakan sel kosong, tetapi dibaca oleh Pandas sebagai missing values
# index 3 tertulis 12 seharusnya missing values karena isi sel seharusnya Y atau N
```

```python
df['OWN_OCCUPIED'].isnull()
```

```python
cnt=0
for row in df['OWN_OCCUPIED']:
    try:
        int(row)
        df.loc[cnt, 'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
```

```python
df.head(9)
```

## Summarizing Missing Values

```python
df.isnull().sum()
```

```python
#quick assessment untuk melihat apakah kita memiliki missing values

df.isnull().values.any()
```

```python
# untuk mengetahui jumlah total missing values

df.isnull().sum().sum()
```

```python
# mengisi missing values dengan nilai median

median = df['NUM_BEDROOMS'].median()
df['NUM_BEDROOMS'].fillna(median, inplace=True)
```
