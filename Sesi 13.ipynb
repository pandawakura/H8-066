# Naive Bayes Classifier

```python
weather = [
    'Sunny',
    'Sunny',
    'Overcast',
    'Rainy',
    'Rainy',
    'Rainy',
    'Overcast',
    'Sunny',
    'Sunny',   
    'Rainy',
    'Sunny',
    'Overcast',
    'Overcast',
    'Rainy'
]
```

```python
temp = [
    'Hot',
    'Hot',
    'Hot',
    'Mild',
    'Cool',
    'Cool',
    'Cool',
    'Mild',
    'Cool',
    'Mild',
    'Mild',
    'Mild',
    'Hot',
    'Mild'
]
```

```python
play = [
    'No',
    'No',
    'Yes',
    'Yes',
    'Yes',
    'No',
    'Yes',
    'No',
    'Yes',
    'Yes',
    'Yes',
    'Yes',
    'Yes',
    'No'
]
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

weather_encoded=le.fit_transform(weather)

print(weather_encoded)
```

```python
temp_encoded=le.fit_transform(temp)
label = le.fit_transform(play)

print( 'temp:',temp_encoded)
print('play:',label)
```

```python
# COMBINING WEATHER AND TEMP INTO SINGLE LIST OF TUPLES

features = list(zip(weather_encoded, temp_encoded))
features
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(features, label)

predicted = model.predict([[0,2]])
print("Predicted Value:", predicted)
```

```python
label_pred =model.predict(features)
label_pred
```

```python
from sklearn import metrics

metrics.accuracy_score(label, label_pred)
```

# Naive Bayes with Multiple Labels

```python
from sklearn import datasets

wine = datasets.load_wine()
```

```python
wine.target_names
```

```python
wine.data.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target,
                                                   test_size=0.3, random_state=109)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)
```

# decision tree classifier

```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
```

```python
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'predigree', 'age', 'label']
 

pima = pd.read_csv("https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/diabetes.csv", header=None, names = col_names,
                  skiprows = 1
                  )
```

```python
pima.head()
```

```python
pima.info()
```

```python
numer = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'predigree', 'label']

for col in numer:
         pima[col] = pd.to_numeric(pima[col], errors='coerce')
```

```python
pima.dropna(inplace=True)
```

```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'predigree']

X = pima[feature_cols]
y = pima.label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier()

clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
# Akurasi

metrics.accuracy_score(y_test, y_pred)
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```
